```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Rainfall_mm': [50, 40, 60, 80, 150, 300, 450, 400, 250, 120, 70, 45],
    'Flood_Incidents': [0, 0, 0, 1, 2, 5, 8, 7, 4, 2, 1, 0]
}
df = pd.DataFrame(data)
```

```python
correlation = df['Rainfall_mm'].corr(df['Flood_Incidents'])
print(f"The Correlation Coefficient is: {correlation:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Rainfall_mm', y='Flood_Incidents', data=df,
            scatter_kws={'s':100, 'color':'blue'}, line_kws={'color':'red'})
```

```python
plt.title('City Rainfall vs. Flood Incidents Analysis', fontsize=14)
plt.xlabel('Monthly Rainfall (mm)', fontsize=12)
plt.ylabel('Number of Flood Incidents', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
plt.show()
```
